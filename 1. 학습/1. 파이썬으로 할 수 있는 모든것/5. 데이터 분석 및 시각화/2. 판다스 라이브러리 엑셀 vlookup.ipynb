## 04. 행과 열의 이름 바꾸기

## 06. 범위로 vlookup 하기

```python
import pandas as pd
```

```python
data1 = ['라', '나', '다', '가', '마']
data2 = [60, 72, 80, 88, 95]

a = pd.DataFrame(data1 , columns=['이름'])
b = pd.DataFrame(data2 , columns=['점수'])
```

```python
b
```

```python
df1 = pd.merge(a, b, left_index=True, right_index=True)
```

```python
df1
```

```python
data3 = [0, 70, 80, 90]
data4 = ['F', 'C', 'B', 'A']

df2 = pd.DataFrame(data3 , columns=['점수']).merge(pd.DataFrame(data4 , columns=['학점']), left_index=True, right_index=True)
```

```python
df2
```

```python
df3 = pd.merge_asof(df1, df2, on='점수', direction='backward')
```

```python
df3['점수'] > 80
```

```python
import pandas as pd
pd.set_option('display.max_rows', 6)
data1 = [['가', 88], ['나', 72], ['다', 80], ['라', 60], ['마', 95]]
data2 = [[90, 'A'], [80, 'B'], [70, 'C'], [0, 'F']]
df1 = pd.DataFrame(data1, columns=['이름', '점수'])
df2 = pd.DataFrame(data2, columns=['점수', '학점'])
df2
```

```python
a = df1.sort_values(by='점수')
b = df2.sort_values(by='점수', ascending=True)
```
