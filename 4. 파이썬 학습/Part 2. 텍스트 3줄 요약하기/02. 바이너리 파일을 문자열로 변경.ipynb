바이너리 파일이란 0과 1을 이용한 2진수 데이터 만으로 인코딩된 파일이다.
데이터를 처리하고 전송하는데 일반적으로 비용이 적게 든다.
텍스트 파일에 비해서 데이터 처리 속도가 빠르다.
데이터 저장공간도 적게 소모된다.

base 64인코딩은 다양한 통신채널을 통해 바이너리 데이터를 안전하게 전송할 수 있다.

## 바이트 문자열 인코딩/디코딩

```python
import base64
```

```python
string = "Life is too short, We need Python."
```

```python
encoded = base64.b64encode(string)
```

```python
# Ascii 인코딩
bstring = string.encode('ascii')
print(bstring)
```

```python
# base64 인코딩
encoded = base64.b64encode(bstring)
```

```python
print(encoded)
```

```python
# base64 디코딩
decoded = base64.decodebytes(encoded)
```

```python
print(decoded)
```

```python
# ascii 디코딩
decoded_str = decoded.decode('ascii')
print(decoded_str)
```

## 이미지 인코딩/디코딩

```python
import base64
```

```python
path = "image.jpg"
```

```python
# 이미지 확인
from PIL import Image
image = Image.open(path)
image
```

```python
# 바이너리 파일 읽기
with open(path, 'rb') as image:
    image = image.read()
```

```python
image
```

```python
# 바이트 정보 확인
from bitstring import BitArray
input_str = '0xff'
c = BitArray(hex=input_str)
c.bin
```

```python
# base64 인코딩
with open(path, 'rb') as image:
    data = image.read()
    encoded = base64.b64encode(data)
    print(encoded)
```

```python
# base64 디코딩
decoded = base64.decodebytes(encoded)
print(decoded)
```

```python
# 이미지 파일로 저장
file = "decoded.png"

with open(file, 'wb') as file:
    file.write(decoded)
```
