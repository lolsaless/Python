## Open, Close

```python
# 텍스트 파일 생성
f = open("새파일.txt", "w")
```

```python
for i in range(1, 11):
    data = "%d번째 줄입니다.\n" % i
    f.write(data)
f.close()
```

```python
# f.close()를 사용하여 파일을 단느것은 좋은 습관이나, 파이썬 2.5부터는 with문을 사용하여 파일을 자동으로 닫도록 하는
# 파일 컨텍스트 관리자를 사용하는 방법이 더 좋다.

with open("새파일.txt", "w") as file:
    for i in range(1, 11):
        data = "%d번째 줄입니다. \n" % i
        file.write(data)
```

```python
with open("text.txt", "w") as file:
    for i in range(1,101):
        data = f"{i}번째 줄입니다.\n"
        file.write(data)
```

```python
with open("text.txt", "r") as file:
    line = file.readline()
    print(line)
```

```python
with open("text.txt", "r") as file:
    lines = file.readlines()
    for i in lines:
        print(i)
```

```python
with open("새파일.txt", "a") as file:
    for i in range(11, 21):
        data = f"{i}번째 줄입니다.\n"
        file.write(data)
```
