### 이미지에 글자 입력하기

```python
a = 0
img = img_basic.copy()
fnt = ImageFont.truetype("image/SCDream3.otf", 10, encoding = "UTF-8")
draw = ImageDraw.Draw(img)
text_list = [
    row1, row2, row3, row4, row5, row6, row7
]

text_list
```

```python
# 25px씩 간격을 띄워서 해당 텍스트를 만들어진 명함 이미지에 담는다.

for unit in text_list:
    a = a + 25
    draw.text((150, a), unit, font = fnt, fill = "black")

img
```
