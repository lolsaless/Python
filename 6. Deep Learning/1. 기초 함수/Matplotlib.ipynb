### figure와 subplot
matplotlib에서 그래프는 figure라는 큰 틀에 ax라는 subplot을 추가해서 만든다.</br>
`plt.show()` 함수가 실행될 때 일어나는 과정을 설명하겠습니다. 이 과정을 이해하려면 먼저 Matplotlib의 구조와 그 작동 방식을 알아야 합니다.

Matplotlib은 크게 두 가지 방식으로 사용될 수 있습니다: 객체 지향 인터페이스와 pyplot 인터페이스. 여러분이 사용한 코드는 이 두 인터페이스가 혼합된 예시입니다.

1. **객체 지향 인터페이스:** `fig = plt.figure()`와 `ax = fig.add_subplot()` 코드는 객체 지향 인터페이스를 사용합니다. 이 방식에서는 명시적으로 Figure 객체 (`fig`)와 Axes 객체 (`ax`)를 생성하고 관리합니다. `fig`는 전체 차트 또는 그림을 나타내며, `ax`는 그 안의 하나의 그래프 또는 좌표축을 나타냅니다.

2. **pyplot 인터페이스:** `plt.show()`는 pyplot 인터페이스의 함수입니다. pyplot 인터페이스는 사용자가 현재 "활성" 상태인 Figure와 Axes를 암묵적으로 관리합니다. 즉, pyplot 인터페이스는 "백그라운드"에서 현재 작업 중인 Figure와 Axes 객체를 추적합니다.

`plt.show()` 함수를 호출하면 Matplotlib는 현재 활성화된 Figure와 그 안의 모든 Axes를 검색하여 화면에 렌더링합니다. 여기서 중요한 점은 `plt.show()`가 현재 활성화된 모든 Figure를 렌더링한다는 것입니다. 따라서, 여러분이 만든 `fig` (및 그 안의 `ax`)는 현재 활성화된 Figure로 간주되어 `plt.show()`에 의해 자동으로 렌더링됩니다.

결론적으로, `ax`를 직접 불러오지 않더라도 `plt.show()` 함수는 현재 활성화된 모든 Figure와 Axes를 렌더링하기 때문에 결과적으로 `fig`와 `ax`가 포함된 그림이 표시되는 것입니다.

```python
fig = plt.figure()
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot()
plt.show()
```

```python
fig = plt.figure(figsize = (10, 5))
ax = fig.add_subplot()
plt.show()
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(2, 2, 1)
ax2 = fig.add_subplot(2, 2, 2)
ax3 = fig.add_subplot(2, 2, 3)
plt.show()
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(221)
ax2 = fig.add_subplot(222)
ax3 = fig.add_subplot(223)
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot()

data = np.array([1, 2, 3])

plt.plot(data)
plt.show()
```

```python
fig = plt.figure()

x = [1, 2, 3]
y1 = [1, 2, 3]
y2 = [3, 2, 1]

ax1 = fig.add_subplot(211)
plt.plot(x, y1)

ax2 = fig.add_subplot(212)
plt.plot(x, y2)

plt.show()
```

```python
fig = plt.figure()

x = [1, 2, 3]
y1 = [1, 2, 3]
y2 = [3, 2, 1]

ax1 = fig.add_subplot(211)
ax2 = fig.add_subplot(212)

ax1.plot(x, y1)
ax2.plot(x, y2)

plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot()
ax.bar([0, 1, 2, 3, 4], [3, 2, 5, 4, 6])
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot()
ax.bar([0, 1, 2, 3, 4], [3, 2, 5, 4, 6], width = 0.3)
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot()
ax.bar([0, 1, 2, 3, 4], [3, 2, 5, 4, 6], bottom = 2)
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot()
ax.bar([0, 1, 2, 3, 4], [3, 2, 5, 4, 6], align = 'edge')
ax.bar()
plt.show()
```
