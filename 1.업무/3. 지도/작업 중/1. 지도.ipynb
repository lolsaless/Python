### 2. 수정 1단계 코드

```python
import pandas as pd
import folium
import random

# 임의 데이터 생성 및 엑셀 저장
data = {
    "Name": ["Parking Lot A", "Nursing Home B", "Kindergarten C", "Postpartum Center D"],
    "Category": ["Indoor Parking", "Nursing Home", "Kindergarten", "Postpartum Care"],
    "Latitude": [37.5665 + random.uniform(-0.1, 0.1) for _ in range(4)],
    "Longitude": [126.978 + random.uniform(-0.1, 0.1) for _ in range(4)],
    "Information": ["Details about A", "Details about B", "Details about C", "Details about D"],
}
df = pd.DataFrame(data)
excel_path = "map_data.xlsx"
df.to_excel(excel_path, index=False)

# 지도 생성
map_center = [37.5665, 126.978]
m = folium.Map(location=map_center, zoom_start=12)

# 카테고리별 색상
category_colors = {
    "Indoor Parking": "blue",
    "Nursing Home": "green",
    "Kindergarten": "red",
    "Postpartum Care": "purple",
}

# JavaScript와 함께 HTML 파일 저장
map_script = """
<script>
    function changeColor(markerId) {
        var marker = document.getElementById(markerId);
        marker.style.color = "black";
        marker.style.background = "black";
    }
</script>
"""

# 데이터를 불러와 마커 추가
for idx, row in df.iterrows():
    popup_content = f"""
    <b>{row['Name']}</b><br>
    Category: {row['Category']}<br>
    Info: {row['Information']}<br><br>
    <button onclick="changeColor('marker-{idx}')">Mark as Done</button>
    """
    popup = folium.Popup(popup_content, max_width=300)

    marker = folium.Marker(
        location=[row["Latitude"], row["Longitude"]],
        popup=popup,
        icon=folium.Icon(color=category_colors[row["Category"]]),
        id=f"marker-{idx}",
    )
    m.add_child(marker)

# HTML 파일 저장
html_file_path = "map_with_markers.html"
with open(html_file_path, "w") as f:
    f.write(m.get_root().render())
    f.write(map_script)
```
